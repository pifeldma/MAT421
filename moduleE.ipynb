```python
import numpy as no
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline

h=.1
x=np.arange(0, 2*np.pi, h)
y=np.cos(x)

forward_difference=np.diff(y)/h

x_difference=x[:-1:]
exact_solution = -np.sin(x_difference)

plt.figure(figsize=(12,8))
plt.plot(x_difference, forward_difference, '--', \
         label = 'Finite difference approximation')
plt.plot(x_difference, exact_solution, \
         label = 'Exact Solution')
plt.legend()
plt.show()

max_error = max(abs(exact_solution - forward_difference))
print(max_error)
```

```python
def derivative(x,y,n):
  if(n==1):
    return np.gradient(y,x)
  else:
    return np.gradient(derivative(x,y,n-1),x)
a=0
b=10
N_points=10000

x=np.linspace(a,b,N_points, dtype=float)
y=np.cos(x, dtype=float)
plt.figure()
plt.plot(x,y,label='f(x)')

n_max=7
for n in range(1, n_max):
  dy=derivative(x,y,n)
  dy[0:n]=dy[n]
  dy[-n:]=dy[-n-1]
  plt.plot(x, dy, label=f"fprime{n}(x)")

plt.legend()
plt.title("function and its derivative")
```
